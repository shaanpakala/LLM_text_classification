```python
def trimWord(word):

    if (len(word) == 0): return ''

    w = word

    if ((w[len(w)-1] == ",")or(w[len(w)-1] == ".")or(w[len(w)-1] == "?")or(w[len(w)-1] == "%")
        or(w[len(w)-1] == "!")or(w[len(w)-1] == "’")or(w[len(w)-1] == ":")or(w[len(w)-1] == ")")or(w[len(w)-1] == ";")
        or(w[len(w)-1]=="\"")or(w[len(w)-1]=="-")or(w[len(w)-1]=="'")or(w[len(w)-1]=="”")or(w[len(w)-1]=="‘")
        or(w[len(w)-1]=="]")or(w[len(w)-1]==">")
        or(w[len(w)-1]=="0")or(w[len(w)-1]=="1")or(w[len(w)-1]=="2")or(w[len(w)-1]=="3")or(w[len(w)-1]=="4")
        or(w[len(w)-1]=="5")or(w[len(w)-1]=="6")or(w[len(w)-1]=="7")or(w[len(w)-1]=="8")or(w[len(w)-1]=="9")
        or(w[len(w)-1]=="[")or(w[len(w)-1]=="]")):
        w = w[:len(w)-1]

    elif((w[0]=="'")or(w[0]=="’")or(w[0]=="\"")or(w[0]=="-")or(w[0]=="“")or(w[0]=="(")or(w[0]=="‘")
        or(w[0]=="[")or(w[0]=="”")):
        w = w[1:]

    elif ((w[len(w)-2:len(w)] == "’s")or(w[len(w)-2:len(w)] == "\'s")or(w[len(w)-2:len(w)] == "\n")):
        w = w[:len(w)-2]

    else: return w

    return(trimWord(w))
```

# Get Wikipedia Data

```python
txt = f'{work_dir}data/text.txt'

if (clear_txt):
  open(txt, "w").close()
  print("Cleared txt file.")

else:
  print("Did not clear txt file.")
```

```python
txtFile = open(txt, "a")

topic_num=0
for topic in topics:

    www = f"https://en.wikipedia.org/wiki/{topic}"

    www = requests.get(www,headers = headers)
    if (www.status_code != 200):
        print(f"\nError with: \n{topic}\n")

    soup = BeautifulSoup(www.text, 'html.parser')

    paragraphs = soup.find('div', class_ = "mw-body-content").find_all('p')
    del soup

    for paragraph in paragraphs:

        text = paragraph.text.split()

        if ((len(text) < 10) or (len(text) > 512)): continue

        for i in range(len(text)):
            text[i] = trimWord(text[i])

        text = ' '.join(text)

        # skip if too many tokens
        if (tokenizer(text, return_tensors="jax", padding=True).input_ids.shape[1] > 512):
          print(f"Skipped a paragraph in {topic}.")
          continue

        txtFile.write(text + "\n--\n--\n")


    txtFile.write("\n*#@$#\n" + topics[topic] + "\n%$#@\n")

    topic_num+=1
    if ((topic_num % 10 == 0) or (topic_num == len(topics))): print(f"{(topic_num)}/{len(topics)} articles done.")

txtFile.close()

print(f"\nSuccessfully updated\n{txt}")
```
